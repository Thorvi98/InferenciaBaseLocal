```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

file_path = '/content/drive/MyDrive/Colab Notebooks/pessoas.csv'

data = pd.read_csv(file_path)

data['genero'] = data['genero'].astype(str)

# Distribuio da Idade
plt.figure(figsize=(12, 6))
sns.histplot(data['idade'], bins=20, kde=True, color="skyblue")
plt.title("Distribuio de Idade")
plt.xlabel("Idade")
plt.ylabel("Frequncia")
plt.show()

# Distribuio da Renda
plt.figure(figsize=(12, 6))
sns.boxplot(x=data['renda'], color="lightgreen")
plt.title("Distribuio de Renda")
plt.xlabel("Renda")
plt.show()

# Renda por Nvel de Educao
plt.figure(figsize=(12, 6))
sns.boxplot(x='educacao', y='renda', data=data, hue='educacao', palette="pastel", dodge=False)
plt.title("Renda por Nvel de Educao")
plt.xlabel("Nvel de Educao")
plt.ylabel("Renda")
plt.xticks(rotation=45)
plt.legend(title="Nvel de Educao", loc="upper right")
plt.show()

# Idade vs Tempo de Emprego (corrigindo a legenda)
plt.figure(figsize=(12, 6))
sns.scatterplot(x='idade', y='tempo_emprego', data=data, hue='genero', palette="Set1")
plt.title("Idade vs Tempo de Emprego")
plt.xlabel("Idade")
plt.ylabel("Tempo de Emprego")
plt.legend(title="Gnero", loc="upper left")
plt.show()

# Distribuio por Gnero e Situao de Aposentadoria
plt.figure(figsize=(12, 6))
sns.countplot(x='situacao_aposentadoria', hue='genero', data=data, palette="viridis")
plt.title("Distribuio por Gnero e Situao de Aposentadoria")
plt.xlabel("Situao de Aposentadoria (0: No, 1: Sim)")
plt.ylabel("Frequncia")
plt.legend(title="Gnero", labels=["Masculino", "Feminino"])
plt.show()
```
