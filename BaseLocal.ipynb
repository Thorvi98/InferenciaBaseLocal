```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

file_path = '/content/drive/MyDrive/Colab Notebooks/pessoas.csv'

data = pd.read_csv(file_path)

data['genero'] = data['genero'].astype(str)

# Distribuição da Idade
plt.figure(figsize=(12, 6))
sns.histplot(data['idade'], bins=20, kde=True, color="skyblue")
plt.title("Distribuição de Idade")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.show()

# Distribuição da Renda
plt.figure(figsize=(12, 6))
sns.boxplot(x=data['renda'], color="lightgreen")
plt.title("Distribuição de Renda")
plt.xlabel("Renda")
plt.show()

# Renda por Nível de Educação
plt.figure(figsize=(12, 6))
sns.boxplot(x='educacao', y='renda', data=data, hue='educacao', palette="pastel", dodge=False)
plt.title("Renda por Nível de Educação")
plt.xlabel("Nível de Educação")
plt.ylabel("Renda")
plt.xticks(rotation=45)
plt.legend(title="Nível de Educação", loc="upper right")
plt.show()

# Idade vs Tempo de Emprego (corrigindo a legenda)
plt.figure(figsize=(12, 6))
sns.scatterplot(x='idade', y='tempo_emprego', data=data, hue='genero', palette="Set1")
plt.title("Idade vs Tempo de Emprego")
plt.xlabel("Idade")
plt.ylabel("Tempo de Emprego")
plt.legend(title="Gênero", loc="upper left")
plt.show()

# Distribuição por Gênero e Situação de Aposentadoria
plt.figure(figsize=(12, 6))
sns.countplot(x='situacao_aposentadoria', hue='genero', data=data, palette="viridis")
plt.title("Distribuição por Gênero e Situação de Aposentadoria")
plt.xlabel("Situação de Aposentadoria (0: Não, 1: Sim)")
plt.ylabel("Frequência")
plt.legend(title="Gênero", labels=["Masculino", "Feminino"])
plt.show()
```
